```python

from platform import python_version
print('Versão da Linguagem Python Usada Neste Jupyter Notebook:', python_version())

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import colorsys

plt.style.use('seaborn-talk')
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

```

```python
data_frame = pd.read_csv('Dados-pesquisa.csv', sep = ',', low_memory=False)
data_frame 
```

```python
data_frame.describe()
```

```python
# bins quantidade de colunas

data_frame.Age.hist(bins = 60)
plt.xlabel("Idade")
plt.ylabel("Número de profissionais")
plt.title("Distribuiçao de Idade")
plt.show()

data_frame.Gender.hist(bins = 5)
plt.xlabel("Sexo")
plt.ylabel("Número de profissionais")
plt.title("Distribuiçao de Sexo nos cursos de dev")
plt.show()

list(data_frame.columns)
```

```python
# Distribuição de programdores por sexo

labels = data_frame.Gender.value_counts().index
num_of_employes = len(data_frame.EmploymentField.value_counts().index)

#lista de cores
listaHSV = [(x*1.0/num_of_employes, 0.5,0.5) for x in range(num_of_employes)]
listaRGB = list(map(lambda x: colorsys.hsv_to_rgb(*x), listaHSV))
fatias, text = plt.pie(data_frame.Gender.value_counts(), colors= listaRGB, startangle= 90)
plt.axes().set_aspect('equal', 'datalim')
plt.legend(fatias, labels, bbox_to_anchor = (1.05, 1))
plt.title('Sexo')
plt.show()
```
