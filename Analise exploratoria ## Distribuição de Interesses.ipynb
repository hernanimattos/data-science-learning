```python
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import colorsys
import warnings

plt.style.use('seaborn-talk')
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
data_frame = pd.read_csv('Dados-Pesquisa.csv', sep=",", low_memory=False)
data_frame
```

```python
list(data_frame)
```

```python
num_job_role_interested = len(data_frame.JobRoleInterest.value_counts().index)
listaHSV = [(x*1.0/num_job_role_interested, 0.5,0.5) for x in range(num_job_role_interested)]
listaRGB = list(map(lambda x: colorsys.hsv_to_rgb(*x), listaHSV))                                                       
labels = data_frame.JobRoleInterest.value_counts().index
colors = ['OliveDrab', 'Orange', 'OrangeRed', 'DarkCyan', 'Salmon', 'Sienna', 'Maroon', 'LightSlateGray', 'DimGray']


# Grafico de pizza
fatias, text = plt.pie(data_frame.JobRoleInterest.value_counts(), colors=listaRGB, startangle= 90 )
                                                                    
plt.axes().set_aspect('equal', 'datalim')  

plt.legend(fatias, labels, bbox_to_anchor =( 1.25, 1))

plt.title('Interesse Profissional')
plt.show()
# data_frame.JobRoleInterest.value_counts()                                                 
                                                                    
                                                                    
```
